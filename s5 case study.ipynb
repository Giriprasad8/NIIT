```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sb
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv(r'C:\Users\girip\Downloads\DS1_C5_S3_BazilHousing_Data_Hackathon.csv')
```

```python
df
```

```python
#task 1 Average cost of living in each potential relocation city

avg_cost_of_living = df.groupby('city')['total (R$)'].mean().reset_index()
plt.figure(figsize=(10, 6))
sb.barplot(x='city', y='total (R$)', data=avg_cost_of_living)
plt.title('Average Cost of Living in Each City')
plt.xlabel('City')
plt.ylabel('Average Total Cost (R$)')
plt.xticks(rotation=45)
plt.show()
```

```python
#task 2 distribution of the number of rooms

plt.figure(figsize=(10, 6))
sb.histplot(df['rooms'], bins=10)
plt.title('Distribution of Number of Rooms')
plt.xlabel('Number of Rooms')
plt.ylabel('Frequency')
plt.show()
```

```python
#task 3 distribution of the parking spaces

plt.figure(figsize=(10, 6))
sb.countplot(x='parking spaces', data=df)
plt.title('Distribution of Parking Spaces')
plt.xlabel('Number of Parking Spaces')
plt.ylabel('Frequency')
plt.show()
```

```python
#task 4 average total cost (R$) for different types of furniture 

avg_cost_by_furniture = df.groupby('furniture')['total (R$)'].mean().reset_index()
plt.figure(figsize=(8, 6))
sb.barplot(x='furniture', y='total (R$)', data=avg_cost_by_furniture)
plt.title('Average Total Cost by Furniture Type')
plt.xlabel('Furniture Type')
plt.ylabel('Average Total Cost (R$)')
plt.show()
```

```python
#task 5 average total cost (R$) for different flooring types

avg_cost_by_floor = df.groupby('floor')['total (R$)'].mean().reset_index()
plt.figure(figsize=(10, 6))
sb.barplot(x='floor', y='total (R$)', data=avg_cost_by_floor)
plt.title('Average Total Cost by Floor Type')
plt.xlabel('Floor Type')
plt.ylabel('Average Total Cost (R$)')
plt.show()
```

```python
#task 6  number of bathrooms vary with the number of rooms

plt.figure(figsize=(10, 6))
sb.scatterplot(data=df, x='rooms', y='bathroom', s=100, alpha=0.7)
plt.title('Number of Bathrooms vs. Number of Rooms')
plt.xlabel('Number of Rooms')
plt.ylabel('Number of Bathrooms')
plt.show()
```

```python
#task 7 average rent amount for properties with and without parking spaces

average_rent_with_parking = df[df['parking spaces'] > 0]['rent amount (R$)'].mean()
average_rent_without_parking = df[df['parking spaces'] == 0]['rent amount (R$)'].mean()
print("Average Rent Amount for Properties with Parking Spaces:", average_rent_with_parking)
print("Average Rent Amount for Properties without Parking Spaces:", average_rent_without_parking)
plt.figure(figsize=(8, 6))
sb.barplot(x=['With Parking Spaces', 'Without Parking Spaces'], y=[average_rent_with_parking, average_rent_without_parking])
plt.title('Average Rent Amount for Properties with and without Parking Spaces')
plt.xlabel('Parking Spaces')
plt.ylabel('Average Rent Amount (R$)')
plt.show()
```

```python
# task 8 acceptance of animals affect the rent amount

average_rent_by_animal = df.groupby('animal')['rent amount (R$)'].mean()
print("Average Rent Amount for Properties that Accept Animals:", average_rent_by_animal['acept'])
print("Average Rent Amount for Properties that Do Not Accept Animals:", average_rent_by_animal['not acept'])
plt.figure(figsize=(8, 6))
sb.barplot(x=average_rent_by_animal.index, y=average_rent_by_animal.values)
plt.title('Average Rent Amount by Acceptance of Animals')
plt.xlabel('Acceptance of Animals')
plt.ylabel('Average Rent Amount (R$)')
plt.show()
```

```python
#task 8 Relationship between total cost and area of the property

df['total_cost'] = df['rent amount (R$)'] + df['property tax (R$)'] + df['fire insurance (R$)']
plt.figure(figsize=(10, 6))
sb.scatterplot(data=df, x='area', y='total_cost')
plt.title('Relationship between Total Cost and Area')
plt.xlabel('Area')
plt.ylabel('Total Cost (R$)')
plt.show()
```

```python
#task 9  Average rent amount based on number of rooms

plt.figure(figsize=(10, 6))
sb.barplot(data=df, x='rooms', y='rent amount (R$)',ci=None)
plt.title('Average Rent Amount Based on Number of Rooms')
plt.xlabel('Number of Rooms')
plt.ylabel('Average Rent Amount (R$)')
plt.show()
```

```python
#task 10 cities suitable for bachelore ,mid-sized family and large family

sb.boxplot(x='city', y='rent amount (R$)', data=df)
plt.title('Distribution of Rental Amounts in Different Cities')
plt.show()
bachelors_df = df[(df['rooms'] >= 1) & (df['rooms'] <= 2) & (df['rent amount (R$)'] <= 1500)]
print("Suitable cities for bachelors:")
print(bachelors_df['city'].unique())
mid_sized_families_df = df[(df['rooms'] >= 3) & (df['rooms'] <= 4) & (df['rent amount (R$)'] <= 5000)]
print("\nSuitable cities for mid-sized families:")
print(mid_sized_families_df['city'].unique())
large_families_df = df[(df['rooms'] >= 4) & (df['rent amount (R$)'] <= 10000)]
print("\nSuitable cities for large families:")
print(large_families_df['city'].unique())
```
