```python
from datetime import date

items = {
    1: {"Name": "Chocolates", "Quantity": 100, "Cost": 30},
    2: {"Name": "Aalo Bujia", "Quantity": 50, "Cost": 50},
    3: {"Name": "Chocolates", "Quantity": 100, "Cost": 30},
    4: {"Name": "Real Fruit", "Quantity": 78, "Cost": 20}
}
sales = {
    1: {"Name": "Chocolates", "Quantity": 0, "Amount": 0},
    2: {"Name": "Aalo Bujia", "Quantity": 0, "Amount": 0},
    3: {"Name": "Chocolates", "Quantity": 0, "Amount": 0},
    4: {"Name": "Real Fruit", "Quantity": 0, "Amount": 0}
}
while True:
    print("\nWELCOME TO THE GROCERY STORE.")
    print("Items Available:")
    print("%-5s %-15s %-10s %-10s" % ("S.No", "Item", "Quantity", "Cost"))
    for key, value in items.items():
        print("%-5d %-15s %-10d %-10.2f" % (key, value["Name"], value["Quantity"], value["Cost"]))
    a = len(items)
    print("\nFill in your order details:")
    for i in range(1, a + 1):
        while True:
            quantity = int(input(f"\nEnter the number of {items[i]['Name']} you want to buy: "))
            if quantity <= items[i]["Quantity"]:
                items[i]["Quantity"] -= quantity
                sales[i]["Quantity"] += quantity
                sales[i]["Amount"] += quantity * items[i]["Cost"]
                break
            else:
                print(f"Sorry, only {items[i]['Quantity']} {items[i]['Name']} available.")
    customer_name = input("\nEnter your name: ")
    delivery_address = input("Enter your address: ")
    distance = float(input("Enter the distance from the store in miles: "))
    print("\nSALES INVOICE")
    print("%-5s %-15s %-10s %-10s" % ("S.No", "Item", "Quantity", "Total Amount"))
    total_amount = 0
    for key, value in sales.items():
        total_amount += value["Amount"]
        print("%-5d %-15s %-10d %-10.2f" % (key, value["Name"], value["Quantity"], value["Amount"]))
    delivery_charge = 0
    if distance <= 5:
        delivery_charge = 10
    elif 5 < distance <= 10:
        delivery_charge = 12
    elif 10 < distance <= 15:
        delivery_charge = 14
    else:
        delivery_charge = 20
    total_bill = total_amount + delivery_charge
    print("\nTotal Amount: $%.2f" % total_amount)
    print("Delivery Charge: $%.2f" % delivery_charge)
    print("\nTotal Bill Amount: $%.2f" % total_bill)
    continue_shopping = input("\nDo you want to continue shopping? (yes/no): ").lower()
    if continue_shopping != 'yes':
        print("\nThank you for shopping with us! visit again! Goodbye.")
        break
```
