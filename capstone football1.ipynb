```python
#Question 1 Evaluate individual player performance metrics such as goals, assists, yellow cards, red cards, and minutes played to identify keycontributors  to team success.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data['date'] = pd.to_datetime(data['date'])
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns
numeric_data = data[numeric_columns]
print(numeric_data.info())
print(numeric_data.describe())

# Visualize player performance metrics
plt.figure(figsize=(12, 8))

# Distribution of goals
plt.subplot(2, 2, 1)
sns.histplot(data=data, x='goals', kde=True, bins=20, color='skyblue')
plt.title('Distribution of Goals')

# Distribution of assists
plt.subplot(2, 2, 2)
sns.histplot(data=data, x='assists', kde=True, bins=20, color='salmon')
plt.title('Distribution of Assists')

# Distribution of yellow cards
plt.subplot(2, 2, 3)
sns.histplot(data=data, x='yellow_cards', kde=True, bins=20, color='gold')
plt.title('Distribution of Yellow Cards')

# Distribution of red cards
plt.subplot(2, 2, 4)
sns.histplot(data=data, x='red_cards', kde=True, bins=10, color='red')
plt.title('Distribution of Red Cards')

plt.tight_layout()
plt.show()

```

```python
#Question 2 Create player profiles based on attributes such as age, position, dominant foot, and market value to identify potential transfertargets or negotiate contract extensions
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
data['date_of_birth'] = pd.to_datetime(data['date_of_birth'])
current_year = datetime.now().year
data['age'] = current_year - data['date_of_birth'].dt.year
plt.figure(figsize=(16, 6))

# Distribution of age
plt.subplot(1, 3, 1)
sns.histplot(data=data, x='age', bins=20, kde=True, color='skyblue')
plt.title('Distribution of Age')

# Distribution of positions
plt.subplot(1, 3, 2)
sns.countplot(data=data, x='sub_position', palette='Set2')
plt.title('Distribution of Positions')
plt.xticks(rotation=45)

# Distribution of dominant foot
plt.subplot(1, 3, 3)
sns.countplot(data=data, x='foot', palette='Set3')
plt.title('Distribution of Dominant Foot')

plt.tight_layout()
plt.show()

# Visualize relationship between age and market value
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='age', y='market_value_in_eur', hue='foot', palette='Set1')
plt.title('Age vs. Market Value')
plt.xlabel('Age')
plt.ylabel('Market Value (in EUR)')
plt.show()
```

```python
#Question 3 Conduct a comparative analysis of key performance indicators between different teams in the league to identify strengths,weaknesses, and areas for improvement.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Grouping by competition_id and aggregating the KPIs
team_stats = data.groupby('competition_id').agg({
    'goals': 'sum',
    'assists': 'sum',
    'yellow_cards': 'sum',
    'red_cards': 'sum',
    # Add more KPIs as needed
}).reset_index()

# Display the aggregated data
print(team_stats)

# Visualizing the KPIs
plt.figure(figsize=(12, 8))

# Example: Goals scored by each team
sns.barplot(x='goals', y='competition_id', data=team_stats.sort_values(by='goals', ascending=False))
plt.title('Goals Scored by Each Team')
plt.xlabel('Total Goals')
plt.ylabel('Team (Competition ID)')
plt.show()


```

```python
#Question 4 Analyze attendance data across different stadiums and matches to understand fan engagement levels and optimize marketingstrategies for ticket sales and fan experiences

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Average attendance per stadium
avg_attendance_per_stadium =data.groupby('stadium')['attendance'].mean().reset_index()

# Total attendance per match
total_attendance_per_match =data.groupby('game_id')['attendance'].sum().reset_index()

# Plot average attendance per stadium
plt.figure(figsize=(12,60))
sns.barplot(x='attendance', y='stadium', data=avg_attendance_per_stadium.sort_values(by='attendance', ascending=False))
plt.title('Average Attendance per Stadium')
plt.xlabel('Average Attendance')
plt.ylabel('Stadium')
plt.xticks(rotation=45)
plt.show()

# Plot total attendance per match
plt.figure(figsize=(12, 8))
sns.barplot(x='attendance', y='game_id', data=total_attendance_per_match.sort_values(by='attendance', ascending=False))
plt.title('Total Attendance per Match')
plt.xlabel('Total Attendance')
plt.ylabel('Match ID')
plt.show()

```

```python
#Question 5 Analyze referee performance metrics such as fouls called, cards issued, and match control to identify patterns and trends in officiating quality and fairness.    

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Average fouls called per referee
avg_fouls_per_referee = data.groupby('referee')[['yellow_cards', 'red_cards']].sum().reset_index()
avg_fouls_per_referee['total_fouls'] = avg_fouls_per_referee['yellow_cards'] + avg_fouls_per_referee['red_cards']

# Plot average fouls called per referee
plt.figure(figsize=(12, 60))
sns.barplot(x='total_fouls', y='referee', data=avg_fouls_per_referee.sort_values(by='total_fouls', ascending=False))
plt.title('Total Fouls per Referee')
plt.xlabel('Total Fouls')
plt.ylabel('Referee')
plt.xticks(rotation=90)
plt.show()
```

```python
#Question 6 Identify optimal substitution strategies based on game situations, player performance, and historical data to maximize teamefficiency and performance.

import pandas as pd
import matplotlib.pyplot as plt

def suggest_substitution(player_performance, game_situation):
    if player_performance['goals'] == 0 and game_situation['minutes_remaining'] > 60:
        return "Consider substituting player for fresh legs"
    else:
        return "No immediate substitution suggested"
player_performance = {'goals': 0} 
game_situation = {'minutes_remaining': 75} 

suggestion = suggest_substitution(player_performance, game_situation)

# Plot the suggested substitution strategy
plt.figure(figsize=(8, 6))
plt.barh(['Suggested Action'], [1], color='green' if 'Consider' in suggestion else 'red')
plt.title('Substitution Strategy')
plt.xlabel('Action')
plt.yticks([])
plt.show()
```

```python
#Question 7 Conduct a detailed analysis of key game events such as goals, assists, and cards to identify momentum shifts, game-changing moments, and influential players.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

events_df = data[['date', 'minute', 'type', 'player_name', 'goals', 'assists', 'yellow_cards', 'red_cards']]
events_df['date'] = pd.to_datetime(events_df['date'])
events_df = events_df.sort_values(by=['date', 'minute'])
plt.figure(figsize=(12, 8))

# Goals over time
plt.subplot(2, 2, 1)
sns.lineplot(x='date', y='goals', data=events_df.groupby('date')['goals'].sum().reset_index())
plt.title('Goals Over Time')

# Assists over time
plt.subplot(2, 2, 2)
sns.lineplot(x='date', y='assists', data=events_df.groupby('date')['assists'].sum().reset_index())
plt.title('Assists Over Time')

# Yellow cards over time
plt.subplot(2, 2, 3)
sns.lineplot(x='date', y='yellow_cards', data=events_df.groupby('date')['yellow_cards'].sum().reset_index())
plt.title('Yellow Cards Over Time')

# Red cards over time
plt.subplot(2, 2, 4)
sns.lineplot(x='date', y='red_cards', data=events_df.groupby('date')['red_cards'].sum().reset_index())
plt.title('Red Cards Over Time')

plt.tight_layout()
plt.show()
```

```python
#Question 10 Monitor player contract expiration dates to proactively manage contract negotiations, extensions, and transfers to minimize therisk of losing key players on free transfers. 

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Extract relevant columns for analysis
relevant_columns = ['player_name', 'contract_expiration_date']
contract_data = data[relevant_columns]

# Convert contract expiration date to datetime format
contract_data['contract_expiration_date'] = pd.to_datetime(contract_data['contract_expiration_date'])

# Count the number of players with contracts expiring in each month
contract_counts = contract_data['contract_expiration_date'].dt.to_period('M').value_counts().sort_index()
plt.figure(figsize=(12, 6))
contract_counts.plot(kind='bar', color='skyblue')
plt.title('Player Contract Expiration Dates')
plt.xlabel('Expiration Month')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.show()
```
