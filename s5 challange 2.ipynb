```python
#task 1 

def cat_num(df):
    cat = []
    num = []
    for col in df.columns:
        if df[col].dtype == 'object':
            cat.append(col)
        else:
            num.append(col)
    return cat, num

cat, num = cat_num(df)
from tabulate import tabulate
print(tabulate({'Category': cat, 'Numeric': num}, headers=['Category', 'Numeric']))
```

```python
#task 2 Bivariate and multivariate analysis on categorical columns

for col in cat:
    sb.countplot(x=col, data=df)
    plt.title(f"Countplot of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Task 3: Bivariate and multivariate analysis on numerical columns

for col in num:
    sb.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```
