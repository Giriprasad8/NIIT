```python
import numpy as np
```

```python
data=np.genfromtxt(r'C:\Users\girip\Downloads\DS1_C4_S1_Shopping_Data_Challenge.csv',delimiter=',',skip_header=True)
```

```python
#task 1 score greater than 80

customer_spending_score_greater_than_80=data[data[:,3]>80]
spending_score_greater_than_80_count=customer_spending_score_greater_than_80.shape[0]
print("the customer who spend greater than 80:",customer_spending_score_greater_than_80)
print("the customer count who spend greater than 80:",spending_score_greater_than_80_count)
```

```python
#task 2 extract the customers who are 20 to 24 years of age

customers_between20_24=data[(data[:,1] >=20) &(data[:,1]<=24)]
print("Data of customers who are between 20 to 24 years of age:",customers_between20_24)
```

```python
#task 3 lowest and highest annual income

lowest_annual_income=np.min(data[:,2])
highest_annual_income=np.max(data[:,2])
print("lowest annual income $:",lowest_annual_income)
print("highest annual income $:",highest_annual_income)
```

```python
# task 4 customer age less than 45years and income>35k and score >75

customer=data[(data[:, 1] < 45) & (data[:, 2] > 35) & (data[:, 3] > 75)]
count=data[(data[:, 1] < 45) & (data[:, 2] > 35) & (data[:, 3] > 75)].shape[0]
print("Details of customers who are less than 45 years of age, have an annual income greater than $35K, and have a spending score greater than 75:",customer)
print("\nNumber of customers count:",count )
```

```python
#task 5 christmas season company wishes $10 to $50

gift_coupons = np.random.randint(10, 51, size=(data.shape[0], 1)) 
data_with_coupons = np.concatenate((data, gift_coupons), axis=1)
print("Details of customers after adding the new column for gift coupons:",data_with_coupons)
```
