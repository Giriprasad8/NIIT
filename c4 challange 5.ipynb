```python
#task 7
dfh = df_1.groupby(['Displacement', 'Make'])[['City_Mileage_km_litre', 'Highway_Mileage_km_litre']].count()
```

```python
dfh
```

```python
df_sorted = dfh.sort_values(by=['City_Mileage_km_litre', 'Highway_Mileage_km_litre'], ascending=False)
```

```python
df0=df_sorted[0:10]
```

```python
import matplotlib.pyplot as plt


# Plotting the line chart
plt.plot(df0['City_Mileage_km_litre'], df0['Displacement'], color='blue', marker='D', linestyle='-')

# Adding labels and title
plt.xlabel('City_Mileage_km_litre')
plt.ylabel('Displacement')
plt.title('Relationship between Displacement and City Mileage')
plt.grid(True)

# Display the plot
plt.show()
```
