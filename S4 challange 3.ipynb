```python
#task 6 customers who are  not married, selfemployed and non-urban area

customers = df[(df['Married'] == 'No') & 
                        (df['Self_Employed'] == 'Yes') & 
                        (df['Property_Area'] != 'Urban') & 
                        (df['Loan_Status'] == 'N')]
gender_count = customers['Gender'].value_counts()
print("Gender-wise count of customers:",gender_count)


```

```python
#task 7  average income of the customers

average_income = df.groupby('Property_Area')['ApplicantIncome'].mean()
print("Average income of customers based on each area:")
print(average_income)
```

```python
#task 8 before and after update of rural and semiurban area

customers = df[(df['Property_Area'].isin(['Rural', 'Semiurban'])) & 
                        (df['ApplicantIncome'] > 4000) & 
                        (df['CoapplicantIncome'] > 1000) & 
                        (df['Loan_Status'] == 'N')]
num_customers = len(customers)
print("Customers who meet the specified criteria:")
print(customers)
print("\nNumber of customers who meet the specified criteria:", num_customers)
df.loc[customers.index, 'Loan_Status'] = 'Y'
updated_customers = df[(df['Property_Area'].isin(['Rural', 'Semiurban'])) &
                                (df['ApplicantIncome'] > 4000) &
                                (df['CoapplicantIncome'] > 1000) &
                                (df['Loan_Status'] == 'N')]
print("\nCustomers who meet the criteria after updating loan status:")
print(customers)
num_customers_updated = len(updated_customers)
print("\nNumber of customers meeting the criteria after updating loan status:", num_customers_updated)
```

```python
#task 9 average income and count of customers who are 3+ Dependents

dependents_customers = df[(df['Dependents'] =='3+') & 
                        (df['Property_Area'] == 'Urban') & 
                        (df['Loan_Status'] == 'Y')]
average_income =dependents_customers['ApplicantIncome'].mean()
num_customers = len(dependents_customers)
print("Average income of customers:", average_income)
print("Number of customers:", num_customers)
```

```python
#task 10 list of customers live in 3+ dependents and avg income

urban_avg_income = df[df['Property_Area'] == 'Urban']['ApplicantIncome'].mean()
customers = df[((df['Property_Area'] == 'Rural') | (df['Property_Area'] == 'Semiurban')) &
                        (df['Dependents'] >='3+') &
                        (df['ApplicantIncome'] > urban_avg_income)]
print("List of customers who live in rural or semiurban")
print(customers)
avg_income_non_urban_y = df[(df['Property_Area'] != 'Urban') & (df['Loan_Status'] == 'Y')]['ApplicantIncome'].mean()
print("\nAverage applicant income of customers from non-urban areas with loan status 'Y':", avg_income_non_urban_y)
```
