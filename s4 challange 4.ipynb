```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv(r'C:\Users\girip\Downloads\DS1_C4_S4_Used_Car_Prices_Data_Challenge.csv')
```

```python
#task 1 drop the rows with missing data

missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)
cleaned_data = df.dropna()
cleaned_data.reset_index(drop=True, inplace=True)
```

```python
#task 2 display the details of potential cars

filtered_data = df[(df['year'] == 2014) & (df['state'] == 'ca')]
sorted_data = filtered_data.sort_values(by='mp', ascending=True)
potential_cars = sorted_data.head(100)
print(potential_cars)
```

```python
#task 3 display the output in a cross table

filtered_data = df[(df['year'] == 2014) & (df['state'] == 'ca')]
cross_table = pd.pivot_table(filtered_data, values='mp', index=['make', 'model'], aggfunc='mean')
cross_table = cross_table.sort_values(by='mp', ascending=True)
print(cross_table)
```

```python
#task 4 display updated data

chevrolet_cars = df[df['make'] == 'Chevrolet'].copy()
chevrolet_cars['discounted_price'] = chevrolet_cars['mp'] * 0.9
print("Chevrolet Cars with Original Prices and Discounted Prices (10% discount):\n")
print(chevrolet_cars[['make', 'model', 'mp', 'discounted_price']])
```

```python
#task 5  average interior colour price

average_prices_by_color = df.groupby('interior')['mp'].mean()
print("Average Price of the 100 Lowest Priced Cars from 2014 by Interior Color:\n")
print(average_prices_by_color)
```

```python
#task 6  preference of whie and black

filtered_data = df[(df['year'] >= 2010) & (df['year'] <= 2015)]
color_counts = filtered_data.groupby(['make', 'color']).size().reset_index(name='count')
print("Color-wise Availability of Cars for Each Manufacturer from 2010-2015:\n")
print(color_counts)
```

```python
#task 7 BMW car  manufactured

bmw_sedans = df[(df['make'] == 'BMW') & 
                          (df['model'] == 'Sedan') & 
                          (df['year'] >= 2010) & 
                          (df['year'] <= 2015) & 
                          (df['condition'] >= 4.5)]
year_wise_max_price = bmw_sedans.groupby(['model', 'year'])['mp'].max()
print(year_wise_max_price)
```

```python
#task 8  top 100 SUVs

filtered_suv = df[(df['year'] >= 2010) &
                  (df['year'] <= 2015) &
                  (df['odometer'] <= 30000)]
sorted_suv = filtered_suv.sort_values(by='mp', ascending=False)
top_100_suv = sorted_suv.head(100)
print("Top 100 SUVs available between 2010 and 2015 with an odometer reading <= 30000, based on prices (descending order):\n")
print(top_100_suv)
```
