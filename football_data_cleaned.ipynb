```python
import pandas as pd

sheet1 = pd.read_excel(r'C:\Users\girip\Downloads\ds_c8_datafiles-main\ds_c8_datafiles-main\appearances.xlsx')
sheet2 = pd.read_excel(r'C:\Users\girip\Downloads\ds_c8_datafiles-main\ds_c8_datafiles-main\game_events.xlsx')
sheet3 = pd.read_excel(r'C:\Users\girip\Downloads\ds_c8_datafiles-main\ds_c8_datafiles-main\game_lineups.xlsx')
sheet4 = pd.read_excel(r'C:\Users\girip\Downloads\ds_c8_datafiles-main\ds_c8_datafiles-main\games.xlsx')
sheet5 = pd.read_excel(r'C:\Users\girip\Downloads\ds_c8_datafiles-main\ds_c8_datafiles-main\players.xlsx')
merged_df = pd.concat([sheet1, sheet2, sheet3, sheet4, sheet5], ignore_index=False)
merged_df.to_excel('merged_sheet.xlsx', index=True)
print("Merged data saved to 'merged_sheet.xlsx'.")
```

```python
#Finding duplicates

print("Shape of DataFrame before removing duplicates:", merged_df.shape)
merged_df.drop_duplicates(inplace=True)
print("Shape of DataFrame after removing duplicates:",merged_df.shape)
```

```python
#Finding null values

null_values =merged_df.isnull().sum() 
print("Null Values in Dataset:")
print(null_values)
```

```python
#Treating null values 

numerical_cols = merged_df.select_dtypes(include='number').columns
for col in numerical_cols:
    merged_df[col].fillna(merged_df[col].mean(), inplace=True)
categorical_cols = merged_df.select_dtypes(exclude='number').columns
for col in categorical_cols:
    merged_df[col].fillna(merged_df[col].mode()[0], inplace=True)

null_values_after_treatment = merged_df.isnull().sum()
print("\nNull Values after Treatment:")
print(null_values_after_treatment)
```

```python
# clean data is saved in a csv file

merged_df.to_csv("football_data_merged_cleaned.csv", index=False)
```
