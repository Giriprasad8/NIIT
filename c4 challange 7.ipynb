```python
#task 1 rreading data frame 
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'C:\Users\User\Downloads\DS1_C4_S7_Car_Data_Challenge (2).csv')
```

```python
df
```

```python
cars_data=df[["Cylinders", "Kerb_Weight", "Make", "Displacement", 
         "Fuel_Tank_Capacity_litre", "City_Mileage_km_litre",
         "Highway_Mileage_km_litre", "Power", "Torque"]]
```

```python
import warnings
warnings.filterwarnings("ignore")
cars_data.dropna(subset=["Cylinders", "Kerb_Weight", "Make", "Displacement", 
                   "Fuel_Tank_Capacity_litre", "City_Mileage_km_litre",
                   "Highway_Mileage_km_litre", "Power", "Torque"], inplace=True)
```

```python
cars_data.isnull().sum()
```

```python
# task 2
cars_data['Kerb_Weight']=cars_data['Kerb_Weight'].str.replace(r'kg','')
```

```python
cars_data['Kerb_Weight']=cars_data['Kerb_Weight'].astype(float)
```

```python
cars_data["Kerb_Weight"]*=2.2
```

```python
cars_data['Power']=cars_data['Power'].str.extract(r'(\d+)')
```

```python
cars_data['Power']=cars_data['Power'].astype(float)
```

```python
cars_data['Torque']=cars_data['Torque'].str.extract(r'(\d+)')
```

```python
cars_data['Torque']=cars_data['Torque'].astype(float)
```

```python
cars_data
```

```python
Mean = []
Median = []
sd = []
CV = []

for i in cars_data.columns:
    if i != 'Make':
        mean = np.mean(cars_data[i])
        Mean.append(mean)
        median = np.median(cars_data[i])
        Median.append(median)
        SD = np.std(cars_data[i])
        sd.append(SD)
        CV.append((SD / mean) * 100)
```

```python
df2 = pd.DataFrame([Mean, Median, sd, CV], columns=['Cylinders', 'Kerb_Weight', 'Displacement', 'Fuel_Tank_Capacity_litre', 'City_Mileage_km_litre', 'Highway_Mileage_km_litre', 'Power', 'Torque'])
df2.insert(0, column='Measure', value=['Mean', 'Median', 'Std. Dev', 'CV'])
```
