```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sb
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv(r'C:\Users\girip\Downloads\DS1_C5_S2_EmployeeAttrition_Data_Practice.csv')
```

```python
df
```

```python
#task 1 Perform bivariate and multivariate analysis on categorical columns 

plt.figure(figsize=(10, 8))

# Analyze business travel category-wise employee attrition
plt.subplot(3, 3, 1)
sb.countplot(x='BusinessTravel', hue='Attrition', data=df)
plt.xticks(rotation=45)

# Analyze education field-wise employee attrition
plt.subplot(3, 3, 2)
sb.countplot(x='EducationField', hue='Attrition', data=df)
plt.xticks(rotation=45)

# Analyze department-wise employee attrition
plt.subplot(3, 3, 3)
sb.countplot(x='Department', hue='Attrition', data=df)
plt.xticks(rotation=45)

# Analyze environment satisfaction score-wise employee attrition
plt.subplot(3, 3, 4)
sb.countplot(x='EnvironmentSatisfaction', hue='Attrition', data=df)

# Gender-wise employee attrition
plt.subplot(3, 3, 5)
sb.countplot(x='Gender', hue='Attrition', data=df)

# Marital status-wise employee attrition
plt.subplot(3, 3, 6)
sb.countplot(x='MaritalStatus', hue='Attrition', data=df)

# Overtime category-wise employee attrition
plt.subplot(3, 3, 7)
df[df['Attrition'] == 'Yes']['OverTime'].value_counts().plot(kind='pie', autopct='%1.1f%%')

# Performance rating-wise employee attrition
plt.subplot(3, 3, 8)
sb.countplot(x='PerformanceRating', hue='Attrition', data=df)

# Job satisfaction category-wise employee attrition
plt.subplot(3, 3, 9)
sb.countplot(x='JobSatisfaction', hue='Attrition', data=df)

plt.tight_layout()
plt.show()
```

```python
#task 2  bi-variate analysis between categorical and numerical variables

plt.figure(figsize=(10, 5))

# Age group-wise distribution of the employee attrition
plt.subplot(1, 2, 1)
sb.histplot(data=df, x='Age', hue='Attrition')
plt.subplot(1, 2, 2)
sb.boxplot(data=df, x='Attrition', y='Age')
plt.tight_layout()
plt.show()
plt.figure(figsize=(15, 5))

# Hourly salary-wise employee attrition
plt.subplot(1, 2, 1)
sb.histplot(data=df, x='HourlyRate', hue='Attrition')
plt.subplot(1, 2, 2)
sb.boxplot(data=df, x='Attrition', y='HourlyRate')
plt.tight_layout()
plt.show()
plt.figure(figsize=(15, 5))

# Monthly salary-wise employee attrition
plt.subplot(1, 2, 1)
sb.histplot(data=df, x='MonthlyIncome', hue='Attrition')
plt.subplot(1, 2, 2)
sb.boxplot(data=df, x='Attrition', y='MonthlyIncome')
plt.tight_layout()
plt.show()
plt.figure(figsize=(15, 5))

# Employee attrition by distance from home
plt.subplot(1, 2, 1)
sb.histplot(data=df, x='DistanceFromHome', hue='Attrition')
plt.subplot(1, 2, 2)
sb.boxplot(data=df, x='Attrition', y='DistanceFromHome')
plt.tight_layout()
plt.show()
plt.figure(figsize=(15, 5))

# Tenure-wise employee attrition
plt.subplot(1, 2, 1)
sb.histplot(data=df, x='TotalWorkingYears', hue='Attrition')
plt.subplot(1, 2, 2)
sb.boxplot(data=df, x='Attrition', y='TotalWorkingYears')
plt.tight_layout()
plt.show()
```

```python
#task 3 Job role-wise count of attrited employees who were travelling rarely

plt.figure(figsize=(8,6))
df[df['BusinessTravel'] == 'Travel_Rarely']['JobRole'].value_counts().plot(kind='bar')
plt.xlabel('Count')
plt.ylabel('Job Role')
plt.title('Job role-wise count of attrited employees who were travelling rarely')
plt.show()
```

```python
#task 4 Job role-wise employee attrition from Life Sciences and Medical education background

plt.figure(figsize=(10,6))
df[df['EducationField'].isin(['Life Sciences', 'Medical'])]['JobRole'].value_counts().plot(kind='barh')
plt.xlabel('Count')
plt.ylabel('Job Role')
plt.title('Job role-wise employee attrition from Life Sciences and Medical education background')
plt.show()
```

```python
#task 5 Gender-wise employee attrition from the sales department

plt.figure(figsize=(10, 6))
sb.countplot(x='Gender', hue='Attrition', data=df[df['Department'] == 'Sales'])
plt.title('Gender-wise employee attrition from the sales department')
plt.show()
```
